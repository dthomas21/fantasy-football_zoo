```python
import pandas as pd
import numpy as np
import math
import string

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
madden = pd.read_excel('../fantasy_football/M20 Player Ratings Spreadsheet.xlsx')
projections = pd.read_csv('ffa_rankings.csv')
```

```python
madden = madden.loc[(madden['position'] == 'HB') | (madden['position'] =='WR') | (madden['position'] == 'QB') | (madden['position'] == 'TE')]
madden['Name'] = madden['Name'].apply(lambda x: x.lower())
projections['player'] = projections['player'].apply(lambda x: x.lower())
madden = madden.reset_index(drop=True)
```

```python
madden['Name'] = madden['Name'].apply(lambda x: x.replace("'", ''))
projections['player'] = projections['player'].apply(lambda x: x.replace("dk metcalf", 'd.k. metcalf'))
```

```python
for i in range(len(madden.Name)):
    if len(madden.Name[i].split()) > 2:
        madden['Name'][i] = madden['Name'][i].split()[0] + ' ' + madden['Name'][i].split()[1]
```

```python
pre_df = pd.merge(madden, projections, 'left', left_on='Name', right_on='player')
df = pre_df[['Name', 'Team', 'position_x', 'ovr rating', 'age_x', 'years pro', 'bye', 'overallRank', 'positionRank']]
```

```python
df['null'] = ''
df['null'] = df['positionRank'].apply(lambda x: 1 if math.isnan(x) == True else 0)
```

```python
df.loc[df.null == 1]
```

```python
df['rookie'] = df['years pro'].apply(lambda x: 0 if x == 0 else 1)
```

```python
df['under_90'] = df['ovr rating'].apply(lambda x: 0 if x < 90 else 1)
```

```python
# assign rookie score
# df.loc[df['rookie']==1]
```

```python
test = df.loc[df['null'] != 1]
test = test.reset_index(drop=True)
```

```python
test['dt_score'] = ''
for i in range(len(test)):
#         print(i)
        test['dt_score'][i] = np.mean([test['rookie'][i], test['under_90'][i]*2, test['overallRank'][i]]) #, test['years pro'][i]])
    
```

```python
test.columns
```

```python
test = test[['Name', 'dt_score', 'overallRank', 'position_x', 'Team', 'bye', 'positionRank', 'overallRank', 'ovr rating', 'years pro', 'age_x']]
```

```python
test = test.sort_values(by='dt_score', ascending=True).reset_index(drop=True)
```
